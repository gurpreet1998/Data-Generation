# Generating data - having particular mean and specific range

```
import numpy as np
import random
```

```
def gen_avg(expected_avg=100, n=1000, a=90, b=109):
  z = [random.randint(a, b) for i in range(n)]
  mean= np.mean(np.sum(z))
  i = 0 
  itr = 10000
  while (expected_avg != mean):
        l = [random.randint(a, b) for i in range(n)]
        mean = np.mean(l)

        if expected_avg == mean:
          print("hurraaah!!!!--")
          return l
        #i = i+1
        
        if i < itr:
          
          print("AGAIN CHECKING-",i)
          i = i+1
          
        else:
          print("NOT FOUND in first %d. searches" %(itr))
          expected_avg -= 0
          itr = itr+5000
          if(itr >10000):
            print("No list found")
            break          
```
